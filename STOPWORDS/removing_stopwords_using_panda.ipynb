```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from spacy.lang.en import English
STOP_WORDS
```

```python
nlp = spacy.load('en_core_web_sm')
doc = nlp("This is a sentence.")
for token in doc:
  if token.is_stop:
    print(token.text)
```

```python
def preprocess(text):
  doc = nlp(text)
  n= [token.text for token in doc if not token.is_stop]
  return " ".join(n)
```

```python
preprocess("This is a sentence.")
```
